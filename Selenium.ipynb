```python
import time
options = webdriver.ChromeOptions() #Chrome or FireFox
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
options.add_argument('--headless')
driver = webdriver.Chrome(executable_path='/Users/pm3796gb/Downloads/chromedriver') #Path of the Webdriver 

driver.get("https://www.imdb.com/title/tt7366338/reviews?ref_=tt_ql_3")
more_buttons = driver.find_elements_by_class_name("ipl-load-more__button") # Name of the buttom 
for i in range (0,130): #For loop to get all the reviews
    for x in range(len(more_buttons)):
            if more_buttons[x].is_displayed():
                driver.execute_script("arguments[0].click();", more_buttons[x])
                time.sleep(1)
page_source = driver.page_source
```

# Texts

```python
soup = BeautifulSoup(page_source)
texts = []
reviews = soup.find_all('div',class_='content') # scope of the data
for review in reviews: #loop into the scope
    review_div = review.find('div', class_='text show-more__control') # next scope 
    if review_div is None:
        review_div = review.find('div', class_='text show-more__control clickable').text #the actual data
    texts.append(review.text.strip()) # add it to the list
```

# Ratings

```python
soup = BeautifulSoup(page_source)
ratings=[]
rating = soup.find_all('div',class_='lister-item-content') # scope of the data
for rate in rating: #loop through the scope
    rate_div = rate.find('div',class_='ipl-ratings-bar') # next scope 
    if rate_div is None:
        rate_div = rate.find('span') #the actual data
    ratings.append(rate.text.strip()) # add it to the list
```

# Titles

```python
soup = BeautifulSoup(page_source)
titles=[]
title = soup.find_all('a',class_='title') # scope of the data
for t in title: #loop through the scope
    title_div = t.find('a',class_="title")# next scope 
    if title_div is None:
        title_div = t.find('a',class_="title") #the actual data
       
    titles.append(t.text.strip()) # add it to the list
```
