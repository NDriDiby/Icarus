```python
decision = other_ter_motion.groupby(["motion_type","decison"]).count().sort_values("mudac_id", ascending=False)
decision.head(10)
```

```python
motion = decision.pivot_table(index= "motion_type", columns= "decison", values="mudac_id")
```

```python
motion["total motion"] = motion.sum(axis=1)
motion.head()
motion.fillna(0)
```

```python
motion.sum(axis=0)
```

```python
motion["Denied_per"] = motion["Denied"]/motion["total motion"]*100
motion["Denied as Moot_per"] = motion["Denied as Moot"]/motion["total motion"]*100
motion["Granted_per"] = motion["Granted"]/motion["total motion"]*100
motion["Granted in Part_per"] = motion["Granted in Part"]/motion["total motion"]*100
motion["Other_per"] = motion["Other"]/motion["total motion"]*100
motion.fillna(0)
```

```python
motion_per = motion.iloc[:,6:]
motion_per = motion_per.fillna(0)
```

```python
motion_per = motion_per.sort_values("Denied_per",ascending=True)
```

```python
motion_per.plot(kind="barh",figsize=(18,12))
plt.xticks(fontsize=15,rotation=0,weight="bold")
plt.xlabel("Percentage",fontsize=14,weight='bold')
plt.ylabel("Other Motion",fontsize=14,weight='bold')
plt.yticks(weight="bold",fontsize=15)
plt.legend(loc="best", fontsize=14)
plt.title(" Ratio of Non-Termination Motion per Decision",fontsize=15)
plt.savefig("Ratio of Non-Termination Motion per Decision.png")
```
